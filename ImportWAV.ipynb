```python
import glob
import scipy.io.wavfile
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
files = glob.glob("C:\\Users\\Travis\\Dropbox\\EScience\\acoustical_monitoring\\rawData\\*_18.wav")
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)
```

```python
plt.figure(figsize=(30,20*len(files)))

for i in range(len(files)):
    a = scipy.io.wavfile.read(files[i])
    
    plt.subplot(len(files),1,i+1)
    plt.plot((1/a[0])*np.arange(0,len(a[1][:,1]),1),a[1][:,1])
    plt.xlabel('Time')
    plt.ylabel('Signal')
    plt.title(files[i][-15:-4])
    plt.grid()
```
